```python
import pandas as pd
import numpy as np
import faiss
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import cosine
from scipy.sparse import csr_matrix
```

```python
df = pd.read_pickle('subredditvecspmi.pkl.gz', compression='gzip')

# Select numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Extract numeric values
data_array = numeric_df.to_numpy(dtype=np.float32)

# Calculate L2 norm for each row
row_norms = np.linalg.norm(data_array, axis=1)

# Divide each row by its L2 norm to normalize the vectors
normalized_data = data_array / row_norms[:, np.newaxis]

# Create a FAISS index using the normalized data
index = faiss.IndexFlatIP(normalized_data.shape[1])
index.add(normalized_data)
```

```python
query_subreddit = "Economics"
query_vector = numeric_df.loc[query_subreddit].to_numpy(dtype=np.float32)
query_vector /= np.linalg.norm(query_vector)

# Perform a cosine similarity search
k = 20  # Number of nearest neighbors to retrieve
distances, indices = index.search(np.array([query_vector], dtype=np.float32), k)

# Get the names of the closest neighbors
closest_neighbors = numeric_df.iloc[indices[0]].index.tolist()
closest_neighbors
```
